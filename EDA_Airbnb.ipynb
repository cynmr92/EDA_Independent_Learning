```python
# Clean the data
# Dealing with missing data
# Check to see the amount of missing values in each column (total and percentage)

Total = listings.isnull().sum().sort_values(ascending = False)
Percent = (listings.isnull().sum()/listings.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([Total,Percent], axis = 1, keys = ['Total', 'Percent'])
missing_data.head(20)
```

##### Since there are many neighbourhoods, we are gonna focus on Manhattan and two within the area that are the most popular: Harlem (Upper Man.) and Tribeca (Lower Man.)

### Manhattan

```python
# Filter out the rows where neighbourhood is Manhattan

prices_manhattan = listings.loc[listings['neighbourhood'] == 'Manhattan', ['price']]
prices_manhattan
```

```python
# Max and min price in Manhattan

min_price_in_manhattan = prices_manhattan['price'].min()
max_price_in_manhattan = prices_manhattan['price'].max()
print(f"Airbnb's in Manhattan are in the range: {min_price_in_manhattan} to {max_price_in_manhattan}")
```

```python
# Mean in Manhattan

mean_manhattan = prices_manhattan['price'].mean()
mean_manhattan
```

```python
x = prices_manhattan['price']

plt.hist(x, density=False, bins=20) 
plt.xlabel('Prices in manhattan');
```

##### Since the last range is too wide, let's try with two neighborhoods within Manhattan: Harlem and Tribeca
